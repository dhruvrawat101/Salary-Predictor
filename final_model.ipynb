```
import numpy as np
import pandas as pd
from sklearn import tree
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
from sklearn.linear_model import LinearRegression
```

```
df1=pd.read_csv('salary.csv')
df1.head()
```

```
dummies=pd.get_dummies(df1.iloc[::,12])
dummies.head(3)
```

```
df2=pd.concat([df1,dummies],axis='columns')
df2.dropna()
```

```
df2.columns
```

```
X = df2[[ 'Rating', 'Age', 'data analyst', 'data engineer', 'data scientist', 'other','python_job', 'excel_job',
       'sql_job', 'tableau_job', ]]

       #removing director ,manager,mle ,job_in_headquarters
       #10 features
X.head()
```

```
Y= df2[['Salary']]
Y.head()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=7)
```

```
#Train the model
from sklearn.ensemble  import RandomForestRegressor
model = RandomForestRegressor(n_estimators=1000)
model.fit(X_train,y_train)
```

```
print("Accuracy --> ", model.score(X_test, y_test)*100)
```

```
model.predict([[3.2,1,1,0,0,1,97,0,0,1]])
```

```
#Saving the model
import pickle
with open('model.pkl','wb') as file:
    pickle.dump(model,file)
```
